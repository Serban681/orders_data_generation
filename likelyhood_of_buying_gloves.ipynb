```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(42)

x_age_group_1_men_1 = np.linspace(18, 45, num=45-17)
y_age_group_1_men_1 = stats.norm.pdf(x_age_group_1_men_1, loc=23, scale=14)
y_age_group_1_men_1 += np.random.normal(0, 0.001, len(x_age_group_1_men_1))

y_age_group_1_scaled_men_1 = y_age_group_1_men_1 * 3300

x_age_group_2_men_1 = np.linspace(45, 75, num=75-44)
y_age_group_2_men_1 = x_age_group_2_men_1 * - 0.5 + 40
y_age_group_2_men_1 += np.random.normal(0, 2, len(x_age_group_2_men_1))

y_age_group_2_scaled_men_1 = y_age_group_2_men_1

x_final_men_1 = np.concatenate((x_age_group_1_men_1, x_age_group_2_men_1))
y_final_men_1 = np.concatenate((y_age_group_1_scaled_men_1, y_age_group_2_scaled_men_1))

x_final_1_men = x_final_men_1.astype(int)
y_final_1_men = np.round(y_final_men_1).astype(int)

# CSV file
import pandas as pd

df = pd.DataFrame({'Age': x_final_1_men, 'Probability': y_final_1_men })
df.to_csv('./likelihood_to_buy/likelihood_men_1.csv')

# Plotting
plt.bar(x_final_1_men, y_final_1_men, color='#457ed9')
plt.title('Likelyhood of buying boxing gloves 1st pair - Men')
plt.xlabel('Age')
plt.ylabel('%')
plt.grid()
```
